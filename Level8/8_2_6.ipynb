### a. ‘Stack’ the table to get it into row form, and demonstrate ‘unstack’ to get it back into the original form.

```python
# Demonstrate STACK (do not change hist)
hist1 = hist.stack(level=0).reset_index().rename(columns = {'level_1':'Ticker'}) # Stack is what will be used as column after Dates. If level = 1, each date, 5 metrics, then next date (change level_1 to 'metrics'). 
del hist1['Adj Close'] # Delete this extra column set

# Remove Na's
hist1 = hist1.dropna()
hist1
```

```python
# Demonstrate UNSTACK (do not change hist)
hist2 = hist.stack(level=0).unstack()
del hist2['Adj Close'] # Delete this extra column set

hist2
```

### b. Melt and unmelt the table (using pivot), as shown in the lecture.

```python
# Demonstrate MELT (do not change hist)
# Yes, it should be the original form. You should use a table that's not a multi index
## Use hist1 above (Non multi-index / single-index. SIMILAR approach to people from QuantNet)
melted = hist1.reset_index().melt(id_vars=['Date', 'Ticker'], var_name="Attribute", value_name="Value") # Can just keep adding to id_vars

# Remove Na's
melted = melted.dropna()
melted
```

```python
# Demonstrate UNMELT using PIVOT (do not change hist)
## 'Should unmelt the one that was melted, similar to the one in the notebook'
pivoted = melted.pivot(index=['Date','Ticker'], columns='Attribute')
pivoted = pivoted['Value'].reset_index()
pivoted.columns.name = None # remove attribute column name

# Remove Na's
pivoted = pivoted.dropna()
pivoted
```

### d. Create a derived table (a.k.a a ‘summary table’), containing the total volume (in millions) per year, of each ticker.

```python
# hist won't work...
# USE hist1 (single index)
# hist1.groupby(['Ticker', 'Date']).agg({'Volume': 'sum'}) # Very very close.
hist1.groupby(['Ticker', hist1['Date'].dt.year]).agg({'Volume': 'sum'}).apply(lambda x:x*1e-6) # Get year for each. DONT FORGET MILLIONS transf
```

### e. Create a derived table containing the total volume (in millions) of each ticker.

```python
# hist won't work...
# USE hist1 (single index)
hist1.groupby(['Ticker']).agg({'Volume': 'sum'}).apply(lambda x:x*1e-6) # Get year for each. DONT FORGET MILLIONS transf
```

### f. Add a column to the original table containing the daily return per ticker (using the above formula (from 8_2_5)).

```python
hist1['Daily Return'] = (hist1.groupby('Ticker')['Close'].apply(pd.Series.pct_change))
# hist1.to_csv('sample.csv') # to check if values are correct. CORRECT.

hist1
```

### g. Add a column to the original table containing the 1-week (5-day) moving average daily return, per ticker.

```python
hist1['1-week (5-day) MA Daily Return'] = hist1.groupby('Ticker')['Daily Return'].transform(lambda x:x.rolling(5).mean())
# hist1.to_csv('sample.csv') # to check if values are correct. CORRECT.

hist1
```

### h. Add a column to the original table containing the rolling cumulative sum of volume per ticker.

```python
# hist1['Rolling Cumulative Volume'] = hist1['Volume'].rolling(5).sum() # Set to 5 as unspecified (Seen in Stack)
# hist1['Rolling Cumulative Volume'] = hist1['Volume'].cumsum() # Use cumsum for simplicity as per lecture
hist1['Rolling Cumulative Volume'] = hist1.groupby('Ticker')['Volume'].transform(lambda x:x.cumsum())
# hist1.to_csv('sample.csv') # to check if values are correct. CORRECT.

hist1
```
