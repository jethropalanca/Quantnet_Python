```python
# import 
import numpy as np
import pandas as pd

# Set up the df 'Per Quantnet, what is descired is all decimals starting from 0.1, increment of 0.1, till 50... (.10, .20, .30... 50)'

# dfSin = pd.DataFrame(np.arange(0, 50.1, 0.1).transpose(), columns = 'Values')
# dfSin.rename(columns={"0": "Values"}, inplace = True) # inplace, need not to re-assign to variable name.
# dfSin = pd.DataFrame(range(0, 50.1, 0.1), columns = 'Values') # range step can only be integer

dfSin = pd.DataFrame([np.arange(0, 50.1, 0.1)]).transpose()
# dfSin.rename(columns={"0": "Values"}, inplace = True) # inplace, need not to re-assign to variable name.
# dfSin.index = ['Values']
dfSin.columns = ['Values']

# Apply sin function
dfSin = dfSin.transform(lambda x: np.sin(x))
dfSin
```

```python
# Plotting
dfSin.plot(figsize=(11,7), grid=True, xlabel = 'Numbers (.1 to 50)', ylabel = 'Sin(x)', color = 'c')
```
