### Part 1. Derive Table

```python
# Derive Table Step 1:
# Part B: 'Overlay year 1 and year 2 histograms.' Use FEW tickers here.
# I'll use data Still from 8_2_6 (similar data set to 9_3)
# I'll use 2 Tickers: FB(1) and C(2)

# Generate Data + Set Up Packages:
# Initialization: Load CSV and Import necessary Packages.

# (1) Packages
import pandas as pd
import yfinance as yf

# (2) Load two years stock data for every ticker in the current S&P 500
# dir(yf)
# Andy Zhang Quantnet Comment (Use read_html):
# REPLACE https://topforeignstocks.com/indices/components-of-the-sp-500-index/. Only has company Name
tableStocks = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
# len(tableStocks) (2 data frames)
# type(tableStocks) (List)

# Extract the PD:
table1 = tableStocks[0] # Get Table 1
tickList = table1['Symbol'] # Get Tickers from Table 1

# Alphabetical (Higher chances of having companies with data on top.)
tickList = sorted(tickList)
tickList

# Load Tickers (Third Attempt Code. Same as Before)
tickers = list(tickList) # Create Ticker List

# New Code to get data:
## Set Start Date/End Date
startDate = '2019-06-30'
endDate = '2021-06-30'

hist = yf.download(tickers, startDate, endDate, group_by='ticker')

## Change: Need set dates only because I will need to get date at the middle for 9_4.
# tickers = yf.Tickers(tickers) # yf.Tickers
# Pull Data (Same with Third Attempt Code. Don't do CSV anymore as file is too large to upload to Quantnet.)
# hist = tickers.history(group_by='ticker', period='2y') # Better.

hist
```

```python
# Derive Table Step 2:
# Single index for use in below questions:
hist1 = hist.stack(level=0).reset_index().rename(columns = {'level_1':'Ticker'}) # Stack is what will be used as column after Dates. If level = 1, each date, 5 metrics, then next date (change level_1 to 'metrics'). 
del hist1['Adj Close'] # Delete this extra column set

# Add Return Column
hist1['Daily Return'] = (hist1.groupby('Ticker')['Close'].apply(pd.Series.pct_change))
hist1

# Drop na's again for stocks with daily return = na on certain days
hist1 = hist1.dropna()
hist1

# Reset Index
hist1 = hist1.reset_index()
del hist1['index'] # Delete this extra column set

# Final Data Table
hist1
```

```python
# Derive Table Step 3:
# Download a table (cross-sectional) containing reference data for each ticker
# Use Table 1 (index 0) from Wikipedia:
crossTable = table1
crossTable = crossTable.sort_values(by=['Symbol']).reset_index() # Sorted
del crossTable['index'] # Delete this extra column set

# Change Security to Company Name
crossTable.rename(columns = {'Security':'Company Name'}, inplace = True)

crossTable
```

```python
# Derive Table Step 4:
# Ten Transactions (Monthly Dollar Cost Averaging during the pandemic):
# Revise: Remove EDP as wikipedia does not have data on limited partnerships. Use KMI instead.
# Revision 2: Add 5 more transactions for better flavor. Current Table is boring. Add personal IPO plays.
transactionsTable = pd.DataFrame()
transactionsTable['Date'] = ['2019-07-31', '2019-08-31', '2019-09-30', '2019-10-31', '2019-11-30', '2019-12-31', '2020-01-31', '2020-02-29', '2020-3-31', '2020-4-30', '2020-5-15', '2020-08-25', '2020-09-01', '2020-09-15', '2020-10-05'] # Date
transactionsTable['Ticker'] = ['AAPL', 'KMI', 'SPGI', 'KMI', 'AAPL', 'AAPL', 'KMI', 'AAPL', 'TSLA', 'SPGI', 'TSLA', 'AAPL', 'AAPL', 'TSLA', 'WFC'] # Build Tech / Safe Portfolio / Own Company Stock. # Change PLTR not a part of Index.
transactionsTable['Amount'] = [1000, 1000, 1000, 1000, -200, -200, -500, 2000, 500, 500, 500, -500, -500, -750, 500] # 'think about the sign of the amount (Increase + , Decrease -)'
## Set sells to (-) as we will aggregate values (i.e. add) to get total position for a particular stock within a given date
transactionsTable['BuySell'] = ['Buy', 'Buy', 'Buy', 'Buy', 'Sell', 'Sell', 'Sell', 'Buy', 'Buy', 'Buy', 'Buy', 'Sell', 'Sell', 'Sell', 'Buy'] # What if I bought during the pandemic crash? What if I sold during trade war tensions in late 2019?

transactionsTable['Date'] = transactionsTable['Date'].transform(lambda x:pd.to_datetime(x)) # change from text to date
# Date = transactionsTable['Date']
# Date[0] # returns a timestamp, is now a date

transactionsTable
```

```python
# Derive Table Step 5:
# Create subdf (open, volume, daily return):
histNew = hist1[['Date','Ticker','Volume','Open','Close','Daily Return']] # Added Close to be able to appreciate Daily Return
histNew = histNew.dropna()

# Further modification to make portfolioSnap work
# histNew = histNew.set_index('Date')
histNew = histNew.reset_index()
del histNew['index'] # Delete this extra column set

histNew
```

```python
# Derive Table Step 6:
def portfolioSnap(date):
    '''
    Takes a date and returns a (cross-sectional) table containing a snapshot of your portfolio at that point in time (Ticker, Position).
    '''
    
    # subset = transactionsTable[transactionsTable['Date'] <= date] (Gets a subset, but need to contain ticker and position)
    # Solution (per Avi Palley: A simpler approach would be for the function in part c to:
    # Calculate the actual current position of each Ticker for the passed-in date. This can be done using a table filter and groupBy)
    
    # FUNCTION
    subset = transactionsTable[transactionsTable['Date'] <= date].groupby(by=['Ticker'], as_index=False).agg({'Amount':'sum'}).rename(columns = {'Amount':'Position'})
    # subset['Date'] = date
    # print(type(date)) str...
    
    # Feature 1:
    # Replaced EPD with KMI (still high div.) as EPD (an LP) was not present in the Wikipedia table
    subset = subset.join(crossTable.set_index('Symbol'), on='Ticker', how='inner') # not right join (get many unnecessary data)

    # Dates (Let Feature 2 intersect with this date):
    subset.insert(loc=0, column='Date', value=pd.to_datetime(date))
    # print(type(pd.to_datetime(date))) Correct!
    
    # Feature 2 (join 'Open', 'Volume' and 'Daily Return' of 8_2_6 table):
    subset = subset.join(histNew.set_index(['Ticker', 'Date']), on=['Ticker', 'Date'], how='inner')
    
    # Error-Trap if dates have an issue.
    if not len(subset):
        raise ValueError('Please Enter a different date because current date (holiday/non-BD) is unable to populate any data.')
    else:
        return subset
    
result = portfolioSnap('2020-04-20')
result
```

```python
# Derive Table Step 7:
# No data if no transactions for those dates... 

import logging

def createScaffold():    
    '''
    Concatenates portfolioSnap for every business date over the two-year period.
    '''
    
    # Create empty data frame
    dateInput = '06-30-2020' # Labels Input (Take labels from portfolioSnap)
    scaffold = pd.DataFrame(columns = list(portfolioSnap(dateInput).columns.values), dtype=float)
    
    # Get List of Business Dates
    dateList = hist1['Date'].unique()
    dateList = pd.bdate_range(dateList[0], dateList[-1]) # bounds are inclusive

    dateList
    
    # Concatenate over that two year Period.
    for date in dateList:
        # print(date). ERROR when unable to populate data for a certain date.
        try:
            scaffold = pd.concat([scaffold, portfolioSnap(date)])
        except:
            logging.warning(f'Unable to populate any data for {date.strftime("%m-%d-%Y")}. Skipping...')
            pass
    
    scaffold = scaffold.reset_index()
    del scaffold['index'] # Delete this extra column set
    
    return scaffold
    
result = createScaffold()
result
```

```python
# Derive Table Step 8:
# Add a column to RESULT of createScaffold (1):
result['Daily P&L'] = result['Daily Return'] * result['Position']
result
```

```python
# Derive Table Step 9:
# Add a column to RESULT of createScaffold (2):
result['Running Cumulative Daily P&L (Stock)'] = result.groupby('Ticker', as_index=False)['Daily P&L'].cumsum() # need dtype=float (stack) to generate result table above else 'DataError: No numeric types to aggregate' cause you get groupby objects
result

# result.to_csv('sample.csv') # Looks Correct.
```

```python
# Derive Table Step 10:
# Add a column to RESULT of createScaffold (3):
# result['Running Cumulative Daily P&L (Portfolio)'] = result.groupby('Date', as_index=False)['Daily P&L'].cumsum()
result['Running Cumulative Daily P&L (Portfolio)'] = result['Daily P&L'].cumsum() # Just cumsum
result

# result.to_csv('sample.csv') # Looks Correct. Captures really well the performance of an investor during the 2020 bull-run (i.e. lost all gains on march, to lose even a lot of the port, but DCA allowed the investor to earn a lot after having bought in March and April)
```

```python
# Derive Table Step 11 (FINAL TABLE):
### summaryTable = pd.DataFrame(columns = ['Ticker','Average Daily PnL','Total PnL','Overall Yield'], dtype=float)
import numpy as np

# (1) Create Ticker List
summaryTable = pd.DataFrame(dtype=float)
summaryTable['Ticker'] = result['Ticker'].unique()
# summaryTable

# (2) Average Daily PnL
# Bad Code:
# result.groupby('Ticker').agg({'Daily P&L': 'sum'}).sum() # equal to total PnL
# result.groupby('Ticker').agg({'Daily P&L': 'sum'}) # Correct: GETS VALUES. BUT NEEDS RESET_INDEX() or cannot use.
# result.groupby('Ticker').agg({'Daily P&L': 'sum'}).reset_index() # Correct: GETS VALUES.

# Set up tempTable (use this for total PnL), step 1 for ave. daily.
tempTable = result.groupby('Ticker').agg({'Daily P&L': 'sum'}).reset_index()
tempTable

# Count dates where ticker (i.e. days with position for each stock).
tempTable3 = result.groupby(['Ticker','Date'])[['Company Name']].count()
tempTable3
tempTable3 = tempTable3.pivot_table(index = ['Ticker'], columns=[], values=['Company Name'], aggfunc=np.sum).reset_index() # also np.mean and others
tempTable3                                                            

# Plug into summaryTable
summaryTable['Average Daily PnL'] = tempTable['Daily P&L'] / tempTable3['Company Name']
summaryTable['Days with position'] = tempTable3['Company Name']
# summaryTable

# (3) Total PnL
summaryTable['Total PnL'] = tempTable['Daily P&L']
# summaryTable

# (4) Overall Yield
# Get average position:
tempTable2 = result.groupby('Ticker').agg({'Position': 'mean'}).apply(lambda x:round(x)).reset_index()
tempTable2

# Calculate Yield
summaryTable['Overall Yield (%)'] = round(100 * summaryTable['Total PnL'] / tempTable2['Position'], 2)
# summaryTable

# DISPLAY TABLE:
summaryTable
```
